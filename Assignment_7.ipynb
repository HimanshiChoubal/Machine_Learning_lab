```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# --- 1. Load and Preprocess Dataset ---
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv"
column_names = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age', 'Outcome']
df = pd.read_csv(url, names=column_names)

X = df.drop('Outcome', axis=1)
y = df['Outcome']

# Scale features for better ANN convergence
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# --- Define the Builder Function ---
def build_and_evaluate(hidden_activation='relu', epochs=50, batch_size=10):
    # Task 1: Building the ANN with 2 hidden layers
    model = Sequential([
        Input(shape=(8,)), # Matches 8 input features
        Dense(12, activation=hidden_activation), # Hidden Layer 1
        Dense(8, activation=hidden_activation),  # Hidden Layer 2
        Dense(1, activation='sigmoid')           # Output Layer (Binary Classification)
    ])

    # Task 2: Compile and Train
    model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
    model.fit(X_train, y_train, epochs=epochs, batch_size=batch_size, verbose=0)

    # Evaluate performance
    loss, accuracy = model.evaluate(X_test, y_test, verbose=0)
    return accuracy

# --- Task 3: Varying Epochs and Batch Sizes ---
print(" Varying Hyperparameters...")
hyper_results = []
configs = [(20, 10), (20, 32), (100, 10), (100, 32)]

for e, b in configs:
    acc = build_and_evaluate(epochs=e, batch_size=b)
    hyper_results.append({'Epochs': e, 'Batch': b, 'Accuracy': acc})

# --- Task 4: Trying Different Activation Functions ---
print(" Activation Functions...")
act_results = []
activations = ['relu', 'tanh', 'sigmoid', 'softmax']

for act in activations:
    acc = build_and_evaluate(hidden_activation=act, epochs=50, batch_size=10)
    act_results.append({'Activation': act, 'Accuracy': acc})

# --- Final Output Formatting ---
print("\n" + "="*30)
print("HYPERPARAMETER RESULTS")
print("="*30)
print(pd.DataFrame(hyper_results))

print("\n" + "="*30)
print("ACTIVATION RESULTS")
print("="*30)
print(pd.DataFrame(act_results))
```
