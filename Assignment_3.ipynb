```python
import pandas as pd
import numpy as np

df = pd.read_csv('Heart - Heart.csv')
print(f"Shape: {df.shape}")
print(f"\nMissing Values:\n{df.isnull().sum()}")
print(f"\nData Types:\n{df.dtypes}")
print(f"\nZeros Count:\n{(df == 0).sum()}")
print(f"\nAge Stats: Mean={df['Age'].mean():.2f}, Median={df['Age'].median()}, Std={df['Age'].std():.2f}")
corr = df.select_dtypes(include=[np.number]).corr()
print(f"\nCorrelation with Age:\n{corr['Age'].sort_values(ascending=False)}")
for col in ['Age', 'RestBP', 'Chol', 'MaxHR', 'Oldpeak']:
    Q1, Q3 = df[col].quantile(0.25), df[col].quantile(0.75)
    IQR = Q3 - Q1
    outliers = df[(df[col] < (Q1 - 1.5 * IQR)) | (df[col] > (Q3 + 1.5 * IQR))]
    print(f"Outliers in {col}: {len(outliers)}")
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
df=pd.read_csv('Heart - Heart.csv')
features=['Age','Sex','ChestPain','RestBP','Chol']
df_subset=df[features].copy()
scaler=StandardScaler()
numerical_cols=['Age','RestBP','Chol']
df_subset[numerical_cols]=scaler.fit_transform(df_subset[numerical_cols])
train_df,test_df=train_test_split(df_subset,test_size=0.25,random_state=42)
print("Extraction Done successfully")
print("Standardized Data:")
print(df_subset.head())
print(f"\n Data Splitting Done Successfully")
print(f"Training set shape: {train_df.shape}")
print(f"Testing set shape: {test_df.shape}")
```

```python
from math import fabs
total=500
actual_positive=50
predicted_positive=100
true_positive=45
false_positive=predicted_positive-true_positive
false_negative=actual_positive-true_positive
actual_negative=total-actual_positive
true_negative=actual_negative-false_positive
accuracy=(true_positive+true_negative)/total
precision=true_positive/(true_positive+false_positive)
recall=true_positive/(true_positive+false_negative)
f1_score=2*(precision*recall)/precision+recall
fpr = false_positive / (false_positive + true_negative)
fnr = false_negative / (true_positive + false_negative)
specificity = true_negative / (false_positive + true_negative)

print("--- Confusion Matrix Components ---")
print(f"True Positives (TP): {true_positive}")
print(f"False Positives (FP): {false_positive}")
print(f"False Negatives (FN): {false_negative}")
print(f"True Negatives (TN): {true_negative}")

print("\n--- Performance Metrics ---")
print(f"I.   Accuracy:    {accuracy:.2f}")
print(f"II.  Precision:   {precision:.2f}")
print(f"III. Recall:      {recall:.2f}")
print(f"IV.  F1-score:    {f1_score:.2f}")
print(f"V.   FPR:         {fpr:.4f}")
print(f"VI.  FNR:         {fnr:.2f}")
print(f"VII. Specificity: {specificity:.4f}")
```
